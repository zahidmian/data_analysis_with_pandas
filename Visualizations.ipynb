```python
import pandas as pd
from pandas_datareader import data
import datetime as dt

import matplotlib.pyplot as plt

# necessary for jupyter
%matplotlib inline 
```

```python
start = dt.datetime(2004,1,1)
end = dt.datetime(2017,12, 31)
goog = data.DataReader("GOOG", 'yahoo', start=start, end=end)
```

```python
goog["High"].plot()
```

```python
goog.plot(y='High')
```

```python
goog.plot(y=["High", "Low"])
```

```python
plt.style.available
```

```python
plt.style.use("fivethirtyeight")
goog.plot(y="Close")
```

```python
plt.style.use("dark_background")
goog.plot(y="Close")
```

#### Bar Charts

```python
def rank_perf(price):
    if price <= 200:
        return "Poor"
    elif price > 200 and price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
    
goog["Close"].apply(rank_perf).value_counts().plot(kind="barh")
```

```python
goog["Close"].apply(rank_perf).value_counts().plot(kind="bar")
```

```python
goog["Close"].apply(rank_perf).value_counts().plot(kind="pie")
```

```python
goog["Close"].mean()
```

```python
def rank_perf2(price):
    if price >= 390.890:
        return "Above Average"
    else:
        return "Below Average"
    
goog["Close"].apply(rank_perf2).value_counts().plot(kind="pie", legend=True)
```
