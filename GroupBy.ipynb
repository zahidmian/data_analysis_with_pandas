## groupby multiple column

```python
sectors = df.groupby(["Sector", "Industry"])
sectors.head()
```

```python
# size shows counts by Sector/Industry combo
sectors.size()
```

```python
sectors.sum()
# returns multi-index dataframe
```

```python
sectors["Revenue"].sum()
```

## .agg() method

```python
sectors.agg({"Revenue": "sum", "Employees": "mean", "Profits": "median"})
```

```python
# mix/match columns and operations
sectors.agg({"Revenue": ["sum", "mean"], "Profits": "sum", "Employees": ["mean", "median", "sum", "min", "max"]})
```

```python
# calculate multiple operations on each column
sectors.agg(["size", "sum", "mean"])
```

## loop through sectors

```python
# determine highest revenue company in each sector

# first create empty dataframe with same columns as original df
dfx = pd.DataFrame(columns=df.columns)

# now loop through each sector and use nlargest to determine which row in the data has the largest Revenue
for sector, data in sectors:
    highest = data.nlargest(1, "Revenue")
    # append row (there's no inplace so reassign it to original variable)
    dfx = dfx.append(highest)

dfx
```

```python
# find company with the most employees by location
cities = df.groupby("Location")
dfx = pd.DataFrame(columns = df.columns)

for city, data in cities:
    highest = data.nlargest(1, "Employees")
    dfx = dfx.append(highest)

dfx
```
