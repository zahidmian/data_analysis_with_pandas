```python
import pandas as pd
```

```python
df =pd.read_csv("data/bigmac.csv", parse_dates=["Date"])
df
```

## .set_index()

```python


# create a mulit-column index
df.set_index(keys=["Date", "Country"], inplace=True)
# notice the nesting 
df.head()
```

```python
# sort the index
df.sort_index(inplace=True, ascending=[True, False])
df
```

```python
# view index
df.index
```

```python
# get the names of the index
df.index.names
```

```python
# extract first element 
df.index[0]
# notice it's a tuple (outer label, inner label)
```

## use index_col parameter to load the file with multi-column index

```python
# another way to set multi-column index
df = pd.read_csv("data/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
df.sort_index(inplace=True)
df.head()
```

## get_level_values()

```python
# get values in index (first element in this case) 
df.index.get_level_values(0)
# same as 
# df.index.get_level_values("Date")
```

## .set_names()
### rename levels

```python
df.index.set_names("Location", level=1)
```

```python
# or simply provide a list of all names
df.index.set_names(["Date", "Location"])
```

```python
df.head()
```

## extract rows from multi index dataframe using tuples

```python
# get all rows that have a Date of 2010-01-01 (first level index)
df.loc[("2010-01-01",)].head()
```

```python
# get rows based on a specific index
df.loc[("2010-01-01", "Ukraine")]
```

```python
# get all rows using second level index
df.loc[pd.IndexSlice[:, 'Ukraine'], :]
```

## .transpose() method

```python
df = pd.read_csv("data/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
df.sort_index(inplace=True)
df = df.transpose()
df
```

```python
df.loc["Price in US Dollars"]
```

```python
# get prices for a specific Date
df.loc["Price in US Dollars", ("2016-01-01",)].head()
```

```python
# get prices for a specific Date/Country
df.loc["Price in US Dollars", ("2016-01-01","Denmark")]
```

```python
# get prices for specific Date, multiple countries
df.loc["Price in US Dollars", ("2016-01-01", ["Denmark", "India"])]
```

## .swaplevel() 

```python
df = pd.read_csv("data/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
df.sort_index(inplace=True)
df.head()
```

```python
df = df.swaplevel()
df.head()
```

```python
df = df.swaplevel(0,1)
df
```

## .stack() Method

```python
df = pd.read_csv("data/worldstats.csv", index_col=["country", "year"])
df.head()
```

```python
x = df.stack()
x.head()
```

```python
# x is a Series object
type(x)
```

```python
x[0:2]
```

```python
x[0]
```

```python
x[2]
```

```python
x["Arab World"]
```

```python
x[("Arab World", 2015)]
```

```python
x[("Arab World", 2015, "Population")]
```

```python
x[("Arab World", 2015, "GDP")]
```

```python
# convert it back to a dataframe
x.to_frame()
```

## .unstack()

```python
x.unstack()
```

```python
# unstack the year index (chaining)
x.unstack().unstack()
```

```python
# 
x.unstack().unstack().unstack()
```

```python
# specify level to unstack the particulr index
x.unstack(level=0)
```

```python
# unstack by name
x.unstack("year")
```

```python
# specify list of indexes
x.unstack(level = [1, 0])
```

```python
x.unstack(level = [0, 1])
```
