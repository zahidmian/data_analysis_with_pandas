## retrieve rows with .loc property (location)
### Access a group of rows and columns by label(s) or a boolean array.

```python
# sorting an index can help with performance
df = pd.read_csv("data/jamesabond.csv", index_col="Film")
df.sort_index(inplace=True)
df.head()
```

```python
# returns series
df.loc["Goldfinger"]
```

```python
# exception if index doesn't exist
try:
    df.loc["Avengers"]
except Exception:
    print("index doesn't exist")
```

```python
# will return all instances of a matching index
df.loc["Casino Royale"]
```

```python
# return sequential values
# between "Dr. No" and "Goldfinger"
df.loc["Dr. No" : "Goldfinger"]
```

```python
# every row from Skyfall onwards
df.loc["Skyfall": ]
```

```python
# multiple index values
df.loc[["Spectre", "Thunderball", "Skyfall"]]
```

```python
# will generate exception if any index doesn't exist
try:
    df.loc[["Spectre", "Avengers", "Skyfall"]]
except Exception:
    print("one of the index values doesn't exist")
```

```python
# use in operator to see if index exists
"Avengers" in df.index
```

## .loc vs iloc


```python
# load dataframe with default index
df = pd.read_csv("data/jamesabond.csv")
df.head()
```

```python
# using iloc to get the 4th row
df.iloc[3]
```

```python
# loc with number works too because the index is numeric
df.loc[3]
```

```python
# slice it in the same way
df.iloc[2:5]
```

```python
# last two rows
df.iloc[-2:]
```

```python
# first two, last two
df.iloc[[1,2, -2, -1]]
```

```python
# get first three columns of first two rows
df.iloc[[1,2], [0,1, 3]]
```

```python
# use callable as well
# get every other row
df.iloc[lambda x: x.index % 2 == 0]
```

```python
# using boolean array
filter_year=df["Year"] > 1990
df[filter_year]
```

## updating values

```python
# change Dr. No to "Dr. Yes"
df.iloc[0,0] = "Dr. Yes"
df.head(2)
```

```python
df.loc[1, "Box Office"] = 449
df.head(2)
```

```python
# update multiple cells
df.loc[1, ["Box Office", "Budget"]] = [449, 13]
df.head(2)
```

## filtering with .where()


```python
df = pd.read_csv("data/jamesabond.csv", index_col="Film")
# notice that rows that don't match
filter_actor = df["Actor"] == "Sean Connery"
df.where(filter_actor)
```

```python
filter_actor = df["Box Office"] > 750
df.where(filter_actor)
```

## filtering with .query()
### column names cannot have spaces

```python
# rename columns to eliminate spaces
df.columns = [c.replace(" ", "_") for c in df.columns]
df.info()
```

```python
df.query('Actor == "Sean Connery" and Director == "Guy Hamilton"')
```

```python
# use str method startswith to do a "like" comparison (contains, endswith, etc.)
df.query('Actor == "Sean Connery" and Director.str.startswith("Guy")')
```

```python
# use not operator
df.query('Actor != "Sean Connery"')
```

```python
# equivalent of "not like"
df.query('Actor.str.contains("Sean")==False')
```

```python
# use in with list to mimic sql in clause
df.query('Actor not in ["Roger Moore", "Daniel Craig"]')
```
